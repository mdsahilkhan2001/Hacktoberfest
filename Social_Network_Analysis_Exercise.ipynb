```python
import networkx as nx
import matplotlib.pyplot as plt

# Load graph
G = nx.read_edgelist('facebook_combined.txt.gz', create_using=nx.Graph(), nodetype=int)

# Calculate degree and plot degree distribution
degrees = [G.degree(n) for n in G.nodes()]
plt.hist(degrees, bins=20)
plt.xlabel('Degree')
plt.ylabel('Number of Nodes')
plt.show()

# Average degree
avg_degree = sum(degrees) / len(degrees)
print("Average degree:", avg_degree)
```

```python
density = nx.density(G)
diameter = nx.diameter(G)
avg_path_length = nx.average_shortest_path_length(G)

print("Network Density:", density)
print("Diameter:", diameter)
print("Average Path Length:", avg_path_length)
```

```python
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
pagerank = nx.pagerank(G)

# Get top influencers
top_influencers = sorted(pagerank, key=pagerank.get, reverse=True)[:10]
print("Top influencers:", top_influencers)
```

```python
import pandas as pd
import networkx as nx

# Load the dataset into a pandas DataFrame
edges = pd.read_csv('/content/facebook_combined.txt.gz', sep=' ', header=None, names=['source', 'target'])

# Create a graph from the edge list
G = nx.from_pandas_edgelist(edges, 'source', 'target')
```

```python
# Check basic information about the graph
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
degrees = [d for n, d in G.degree()]

print(f"Number of nodes (users): {num_nodes}")
print(f"Number of edges (connections): {num_edges}")
print(f"Average degree (connections per user): {sum(degrees) / len(degrees)}")
```

```python
import matplotlib.pyplot as plt

# Get degree for each node
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)

# Plot degree distribution
plt.figure(figsize=(10, 6))
plt.hist(degree_sequence, bins=100)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()

# Average degree
avg_degree = sum(degree_sequence) / len(degree_sequence)
print(f"Average number of connections per user: {avg_degree}")
```

```python
# Network density
density = nx.density(G)
print(f"Network density: {density}")

# Diameter (use approximation for large graphs)
try:
    diameter = nx.diameter(G)
except nx.NetworkXError:
    diameter = nx.diameter(nx.connected_component_subgraphs(G)[0])
print(f"Diameter of the network: {diameter}")

# Average path length
avg_path_length = nx.average_shortest_path_length(G)
print(f"Average path length: {avg_path_length}")
```

```python
# Degree centrality
degree_centrality = nx.degree_centrality(G)

# Betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# PageRank
pagerank = nx.pagerank(G)

# Sort centralities to find top influencers
top_influencers = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10]
print("Top 10 influencers based on degree centrality:")
for influencer in top_influencers:
    print(influencer)
```

```python
# Apply Louvain community detection
partition = community_louvain.best_partition(G)

# Visualize communities
import matplotlib.pyplot as plt
import networkx as nx

# Draw graph with node colors corresponding to their partition
pos = nx.spring_layout(G)  # Positioning for visualization
plt.figure(figsize=(12, 12))
nx.draw_networkx_nodes(G, pos, node_size=30, cmap=plt.cm.RdYlBu,
                       node_color=[partition[n] for n in G.nodes()])
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.title("Communities Detected by Louvain Algorithm")
plt.show()
```

```python
# Simple recommendation system based on common neighbors
def recommend_connections(G, node, k=5):
    # Get all neighbors
    neighbors = set(G.neighbors(node))

    # Get potential new connections by common neighbors
    scores = {}
    for n in G.nodes():
        if n not in neighbors and n != node:
            common_neighbors = len(list(nx.common_neighbors(G, node, n)))
            scores[n] = common_neighbors

    # Recommend top k new connections
    recommendations = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:k]
    return recommendations

user_id = 10  # Replace with any node
print(f"Top 5 recommendations for user {user_id}:")
print(recommend_connections(G, user_id))
```
